### Lists

If you want to learn more about Python lists, check out https://www.tutorialspoint.com/python/python_lists.htm

# Loops
A couple of example on how to use loops. For more info see https://www.tutorialspoint.com/python/python_for_loop.htm or Google.

````python
list_of_numbers = [1.,2.,3.,4.,5.,4.,3.,2.,1.]

incremented_list_of_numbers = []
for i in range(len(list_of_numbers)):
    number = list_of_numbers[i]
    incremented_list_of_numbers.append(number+1)
print('Incremented list:')
print(incremented_list_of_numbers)

#More elegantly
incremented_list_of_numbers2 = []
for number in list_of_numbers:
    incremented_list_of_numbers2.append(number+1)
print('Second incremented list:')
print(incremented_list_of_numbers2)

#We can express the for-loop above as a one-liner using list comprehensions.
# Thus, most elegantly:
incremented_list_of_numbers3 = [number + 1 for number in list_of_numbers]
print('Third incremented list:')
print(incremented_list_of_numbers3)

#looping over dictionaries
for key in d:
    value = d[key]
    print(key,value)
````

### Task 12 [10%]:
Given a vector $v \in \mathbb{R}^n$, create a matrix $ M=\begin{bmatrix} \dots 0&0&0&v_1 \\ \dots 0&0&v_1&v_2 \\ \vdots & \vdots & \vdots & \vdots\\ v_1& v_{2}& \dots & v_{n} \end{bmatrix}$

````python
v = np.arange(5)+1
# Your code goes here:
def vec2matr (vec):
    matr=[]
    for i in range(len(vec)):
        r=np.append(np.zeros(len(vec)-i-1),(vec[:i+1]))
        matr=np.append(matr,r)
    matr=np.reshape(matr, (len(vec), len(vec)))
    matr=np.matrix(matr)
    print(matr)
vec2matr(v)
````

## Conditional Probabilities

### Task 13 [10%]:
Given a vector $v \in \{0,1,2,3\}$, first count the number of occurences of $v_i = k$ given that $v_{i-1} = j$. Then transform the counts into probabilities, i.e. estimate $P(v_i = k|v_{i-1}=j)$. A dictionary might be a suitable data structure.

````python
vt = 0
v = []
choices = {0:[0.1,0.3,0.1,0.5], 1:[0.2,0.3,0.4,0.1], 2:[0,0,0,1], 3:[1,0,0,0]}
for i in range(1000):
    vt = np.random.choice([0,1,2,3], p=choices[vt])
    v.append(vt)
v = np.array(v)
# Your code goes here
v_1 = np.reshape(v,(-1,2))
v_2 = np.reshape(v[1:-1],(-1,2))
#print(v_1.shape)
#print(v_2.shape)
unique_1, counts_1 = np.unique(v_1, axis=0, return_counts=True)
unique_2, counts_2 = np.unique(v_2, axis=0, return_counts=True)
dict_1={}
for i in range(len(unique_1)):
    dict_1[str(unique_1[i])]=counts_1[i]
dict_2={}
for i in range(len(unique_2)):
    dict_2[str(unique_2[i])]=counts_2[i]
#print(dict_1)
#print(dict_2)
dict_sum={}
dict_sum[0]=dict_1["[0 0]"]+dict_1["[0 1]"]+dict_1["[0 2]"]+dict_1["[0 3]"]+dict_2["[0 0]"]+dict_2["[0 1]"]+dict_2["[0 2]"]+dict_2["[0 3]"]
dict_sum[1]=dict_1["[1 0]"]+dict_1["[1 1]"]+dict_1["[1 2]"]+dict_1["[1 3]"]+dict_2["[1 0]"]+dict_2["[1 1]"]+dict_2["[1 2]"]+dict_2["[1 3]"]
dict_sum[2]=dict_1["[2 3]"]+dict_2["[2 3]"]
dict_sum[3]=dict_1["[3 0]"]+dict_2["[3 0]"]
#print(dict_sum)
dict_result={}
dict_result[0]=[(dict_1['[0 0]']+dict_2["[0 0]"])/dict_sum[0],(dict_1["[0 1]"]+dict_2["[0 1]"])/dict_sum[0],(dict_1["[0 2]"]+dict_2["[0 2]"])/dict_sum[0],(dict_1["[0 3]"]+dict_2["[0 3]"])/dict_sum[0]]
dict_result[1]=[(dict_1['[1 0]']+dict_2['[1 0]'])/dict_sum[1],(dict_1['[1 1]']+dict_2['[1 1]'])/dict_sum[1],(dict_1['[1 2]']+dict_2['[1 2]'])/dict_sum[1],(dict_1['[1 3]']+dict_2['[1 3]'])/dict_sum[1]]
dict_result[2]=[0,0,0,(dict_1['[2 3]']+dict_2['[2 3]'])/dict_sum[2]]
dict_result[3]=[(dict_1['[3 0]']+dict_2['[3 0]'])/dict_sum[3],0,0,0]
print(dict_result)
````

### Task 14 [10%]:
Extract a matrix $sundays$ that contains all measurements recorded on a Sunday. Do the same for Mondays and Wednesdays. Plot the mean measurement for those days of the week.

````python
x = np.sin(np.linspace(-np.pi,np.pi, 100))
x = np.outer(x,np.sin(np.linspace(-np.pi*1000./7,np.pi*1000./7, 1000)))
x[:,[i for i in range(1000) if i % 7 == 1]] += np.random.normal(0,1,(100,143))
# Your code goes here
sundays=np.matrix(x[:,[i for i in range(1000) if i % 7 == 0]])
sundays_mean=[]
sundays_mean= [np.append(sundays_mean,np.mean(sundays[i,:])) for i in range(len(sundays))]
plt.plot(sundays_mean,color="blue",label="Sunday")
mondays=np.matrix(x[:,[i for i in range(1000) if i % 7 == 1]])
mondays_mean=[]
mondays_mean= [np.append(mondays_mean,np.mean(mondays[i,:])) for i in range(len(mondays))]
plt.plot(mondays_mean,color='green',label='Monday')
wednesdays=np.matrix(x[:,[i for i in range(1000) if i % 7 == 3]])
wednesdays_mean=[]
wednesdays_mean= [np.append(wednesdays_mean,np.mean(wednesdays[i,:])) for i in range(len(wednesdays))]
plt.plot(wednesdays_mean,color='red',label="Wednesday")
plt.legend(loc='upper right')
````
